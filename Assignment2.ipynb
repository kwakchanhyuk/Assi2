```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm
%matplotlib inline

df = pd.read_csv('C:/Users// /Number.csv',encoding='euc-kr')
a = df.groupby('By administrative divisions')[['Number of establishment (EA)','Number of workers (In Person)']].mean()
b = df.groupby('By industry')[['Number of establishment (EA)','Number of workers (In Person)']].mean()
ax = a.plot(kind ='bar',title='Number of workers and establishment by administartive', fontsize=12)
ax = b.plot(kind ='bar',title='Number of workers and establishment by industry', fontsize=12)
```
